### Predicting price of stock using Random Forests

The independent variable (X_train) takes a subset of the dataset which only contains float values, so we drop the $Date$ and $Company$ columns for each company stock. 


### Amazon

```python
# training set X
def valid(col):
    return 'Date' not in col and 'Company' not in col

X_AMZN = dataAmazon[[x for x in dataAmazon.columns if valid(x)]].iloc[0:-800]

# shifting 800 rows
y_AMZN = dataAmazon.Close.shift(-800).dropna() # removing NA

X_AMZN.info()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
# splitting data into test and training sets
X_train_AMZN, X_test_AMZN, y_train_AMZN, y_test_AMZN = train_test_split(X_AMZN, y_AMZN, random_state=42)
```

```python
# creating a model using Random Forest Regressor
rf_model = RandomForestRegressor()

rf_model.fit(X_train_AMZN, y_train_AMZN)
```

```python
# getting score for Random Forest Regressor model for Amazon stock dataset
rf_model.score(X_test_AMZN, y_test_AMZN)
```

Amazon stock prediction model gives a score of 96%! 

```python
# predicted values
y_predicted_AMZN = rf_model.predict(X_test_AMZN)
y_predicted_AMZN[0:20]
```

```python
# plot of predicted values
plt.plot(y_predicted_AMZN)
```
